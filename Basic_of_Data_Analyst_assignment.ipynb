
** Q1.Define the role of a Data Analyst in your own words. What value do they bring to an organization?**

- A Data Analyst is like a translator between raw data and actionable insight,Their role is to sift through large volumes of information—numbers, trends, patterns—and turn it into something meaningful that helps organizations make smarter decisions.

Value to an Organization:

Improves efficiency by identifying bottlenecks or areas for optimization.

Reduces risk through predictive analysis and forecasting.

Drives strategy by uncovering customer behavior, market trends, and performance metrics.

Enhances profitability by spotting opportunities for growth or cost savings.

** Q4. What is Prompt Engineering, and explain any two types of prompts used in Generative AI?**

- Prompt engineering is the skill of writing clear and effective instructions for AI tools—especially those that generate text, images, or code. It’s all about knowing how to “talk” to the AI so it gives you the best possible result. The better the prompt, the better the output.Instructional prompt & Shot prompt are the two types of prompt.
