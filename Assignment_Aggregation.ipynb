## Use “ world” Database to solve the following questions

##Question 1 : Count how many cities are there in each country?
use World;

SELECT Country_Name, COUNT(distinct City_name) No_of_City

FROM Country c

INNER JOIN city ci

ON c.Country_Code=ci.Countrycode

GROUP BY Country_Name

ORDER BY  No_of_City;


##Question 2 : Display all continents having more than 30 countries.
use world;

SELECT

Continent

FROM Country

Group BY Continent

 Having count( Country_Name)>30;


##Question 3 : List regions whose total population exceeds 200 million.
use world;

Select region, Country_Population/1000000 as Population_in_Mil  from Country

Where Country_Population /1000000>200  

Order by Region;


##Question 4 : Find the top 5 continents by average GNP per country.
Select

Continent, AVG (GNP)  as AVG_GNP

from Country

Where GNP>0

GROUP BY Continent

ORDER BY AVG_GNP DESC

LIMIT 5;

##Question 5 : Find the total number of official languages spoken in each continent.

SELECT

Continent, COUNT(DISTINCT Language) as total_official_language

from Country c

INNER JOIN countrylanguage cl

ON c.Country_Code=cl.CountryCode

WHERE IsOfficial='T'

group By Continent

order by total_official_language DESC;

##Question 6 : Find the maximum and minimum GNP for each continent.

SELECT

Continent, max(GNP) as max_GNP FROM Country

group by Continent

order by max_GNP DESC;

SELECT

Continent, min(GNP) as min_GNP FROM Country

group by Continent

order by min_GNP ;

##Question 7 : Find the country with the highest average city population.

Country_Name, AVG (City_Population) as avg_citypop

on c.country_code=ci.countrycode

##Question 8 : List continents where the average city population is greater than 200,000.

SELECT

continent,

AVG(City_population) as avg_city_pop

FROM Country c

INNER JOIN  city ci

ON c.country_code=ci.countrycode

GROUP BY continent

HAVING avg_city_pop> 200000

order by avg_city_pop DESC;


##Question 9 : Find the total population and average life expectancy for each continent, ordered by average life expectancy descending.

Select

continent,

sum(Country_population) as total_country_pop,

avg(LifeExpectancy) as AVG_lifeExpectancy
