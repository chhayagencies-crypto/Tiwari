Explanation

The RAND() function generates numbers from a continuous uniform distribution between 0 and 1.

That means every number in the interval [0,1] is equally likely.

The histogram should look fairly flat across bins.

With only 100 values, you’ll see random fluctuations — some bins taller, some shorter.

If you increase to 1000 or 10,000 values, the histogram will smooth out and look more evenly flat, clearly showing the uniform distribution.

Q5. What is Kurtosis and their type?

A5. Kurtosis is a statistical measure based on the fourth central moment of a distribution.

It describes whether data are heavy-tailed (more extreme values) or light-tailed (fewer extreme values) compared to the normal distribution.

Mesokurtic, Leptokurtic,Platykurtic are the type of Kurtosis.
