Dataset Overview

  This dataset represents fruit sales across different regions within the first 6 months of 2024.

Columns Explanation:

Date → Transaction date (Jan–Jun 2024).

Region → Sales region (East, West, North, South).

Product → Type of fruit sold (Apples, Oranges, Bananas, Mangoes).

Sales → Total sales amount (₹).

Quantity → Units sold.

Tools for Analysis:

Excel → PivotTables, Conditional Formatting, Charts, Dashboards.

Q4. Explain the difference between Sorting and Filtering with an example from this dataset.

Ans.4  Sorting means arranging data in specific order ascending or descending (String case) or Smallest/largest( Number )

EXample: sort by "Sales" in descending order, the highest sale (₹497) will appear at the top, and the lowest (₹53) at the bottom.


Filter: Filtering means displaying only the rows that meet specific criteria, hiding the rest temporarily.

Example: filter for "Region = South", you'll only see rows where the region is South — all other regions (West, East, North) will be hidden.
