Question 1 : Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?

Ans 1.  In short, Data Quality is the measure of how "fit for use" your data is based on its accuracy, completeness, and reliability.

While Data Cleaning is the manual act of fixing errors (like correcting typos), Data Quality is the entire systemic process of ensuring data stays healthy throughout the ETL journey.

Cleaning makes data look "pretty," but Quality ensures it is meaningful. A pipeline can load a perfectly "clean" list of numbers, but if those numbers represent the wrong currency or are three days late, the data is high-quality in form but low-quality in function.

consistency, timliness, validty , accuracy and completeness.

Question 3 : What is duplicate data? Explain three causes in ETL pipelines.

Ans 3.  Duplicate data is the existence of identical records for the same entity (e.g., two rows for the same sale). It causes "double-counting," leading to inflated numbers and skewed analytics.

1. If a pipeline fails and restarts without "cleaning up" its previous partial run, it loads the same data twice.
2. Using a JOIN on non-unique keys in the Transform stage. If one "Customer" matches three "Addresses," you suddenly have three "Customers" in your output.
3. Merging data from two sources (e.g., CRM and Billing) where the same person is registered with slightly different info, creating two unique-looking rows for one person.



Question 4 : Differentiate between exact, partial, and fuzzy duplicates.

Ans4. n data management, duplicates aren't always identical copies; they often hide under slight variations. Here is how they differ:

1. Exact Duplicates: Every single character and value in a record is an identical match to another record.

2. Partial Duplicates:  The records share the same unique identifier, but other non-essential data might differ (like a timestamp or a status).

3. Fuzzy Duplicates: The records refer to the same real-world entity but contain typos, formatting differences, or missing information. They "look" different to a computer but are the same to a human.


Question 5 : Why should data validation be performed during transformation rather than after loading?

Ans5.Validating during transformation is proactive; validating after loading is reactive.

Prevents "Data Poisoning": It acts as a gatekeeper, stopping bad data from ever reaching your production dashboards.

Saves Money: You avoid the storage and processing costs of "loading junk" into your data warehouse.

Faster Debugging: Itâ€™s easier to catch an error while the data is in flight than to "hunt and peck" through a table of millions of rows later.

Ensures Uptime: Many loads will simply crash if they hit a constraint violation (like a null value in a required column). Validation ensures the load actually finishes.

Question 7 : Write an SQL query on  to list all duplicate keys and their counts using the
business key (Customer_ID + Product_ID + Txn_Date + Txn_Amount )


ANS 7.


SELECT
    customer_id,
    product_id,
    Txn_date,
    Txn_amount,
    COUNT(*) as duplicate_count
FROM sale_datasets
GROUP BY
    customer_id,
    product_id,
    Txn_date,
    Txn_amount
HAVING COUNT(*) > 1;
