A3. Created the new measure of AOV by formula AOV = DIVIDE(SUM(amazon_store_sales_2022[total_discounted_Price]),DISTINCTCOUNT(amazon_store_sales_2022[OrderID]))

Q8.  Create a measure called "Date Difference" that calculates the number of days between the latest  Order Date and today's date using the DATEDIFF function.

Ans8.  Created this measure by
