Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?

A1. A Common Table Expression (CTE) is a temporary, named result set defined within the execution scope of a single SQL statement. It is created using the WITH keyword and can be referenced just like a table or view within the query.

Instead of writing deeply complex subqueries, we can define intermediate results in a clear, step-by-step manner.


Q2. Why are some views updatable while others are read-only? Explain with an example.

A view can be updated if:
It comes from one table only.
It shows the data directly (no totals, no grouping).
Every row in the view matches exactly one row in the table.

A view is read-only if:

It uses GROUP BY or totals (like SUM, COUNT).

Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?

A3. A stored procedure is like a saved set of SQL commands stored inside the database. Instead of writing the same SQL query again and again, you can just call the procedure by name.Stored procedures = faster, safer, reusable, and easier to manage compared to writing raw SQL repeatedly.


Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.

A4. A trigger is a special kind of stored program in a database that automatically runs when certain events happen on a table or view. These events can be:

INSERT (adding new data)

UPDATE (changing existing data)

DELETE (removing data)

Suppose you want to keep track of every time an employee's salary changes. A trigger can automatically insert a record into an audit table whenever the Salary column is updated.

Q5. Explain the need for data modelling and normalization when designing a database

A5. Data modelling is like making a blueprint of your database before you build it.

It shows how tables, columns, and relationships will be organized.

Helps you understand what data you need, how it connects, and how to store it efficiently.

Why it's needed:

Avoids confusion later.

Makes sure the database supports the business needs.

Helps developers and analysts work with the same clear structure.

Normalization is the process of organizing data to reduce duplication and improve consistency.

It breaks big tables into smaller ones and links them with relationships (usually using keys).

Keeps data accurate and consistent (no conflicting values).

Saves storage space.
