Question 1: What is a null hypothesis (H₀) and why is it important in hypothesis testing?

Ans1.  It’s a starting assumption in statistics.

It says: “Nothing special is happening. There’s no effect, no difference.”

Example: If you test a new medicine, H₀ says “This medicine works the same as no medicine.”

it is important as below

Baseline check: It gives us something to test against.

Avoids false claims: Makes sure we don’t say something works when it’s just luck.

Clear decision rule: If the data shows strong evidence against H₀, we reject it and say the alternative (H₁) might be true.


Question 2: What does the significance level (α) represent in hypothesis testing?

Ans2.  The significance level, written as α (alpha), is a threshold we set before doing a hypothesis test.It tells us how much risk we are willing to take of wrongly rejecting the null hypothesis (H₀) when it is actually true.

In other words, α is the probability of making a “false alarm” (also called a Type I error).

Question 5: A company claims that the average time to resolve a customer complaint is 10 minutes.
A random sample of 9 complaints gives an average time of 12 minutes and a standard deviation of 3 minutes.
At α = 0.05, test the claim.

Ans5.  To test the company's claim that the average time to resolve a customer complaint is 10 minutes, we set up a two-tailed hypothesis test at the 5% significance level. The null hypothesis states that the mean resolution time is 10 minutes, while the alternative suggests it is different. From a sample of 9 complaints, the mean time was 12 minutes with a standard deviation of 3 minutes. Using the t-test, the test statistic is calculated as

t=(12-10)/(3/square root of 9)= 2 with 8 degrees of freedom. The critical value at α=0.05 is approximately 2.306. Since the observed
t-value (2) is less than the critical value, and the p-value (≈ 0.078) is greater than 0.05, we fail to reject the null hypothesis. Therefore, the evidence is insufficient to conclude that the true average resolution time differs from 10 minutes, meaning the company's claim stands at the 5% level of significance.

Question 7: The productivity of 6 employees was measured before and after a training program.


image.png



At α = 0.05, test if the training improved productivity.


Ans7.  # Paired t-test for productivity improvement


## Step 1: Hypotheses
- H0: The training has no effect (μd = 0)  
- H1: The training improves productivity (μd > 0)

## Step 2: Calculations
- Mean of differences: d̄ = 3  
- Standard deviation of differences: sd ≈ 1.67  
- Sample size: n = 6  
- Degrees of freedom: df = 5  

Test statistic:  
t = d̄ / (sd / √n) = 3 / (1.67 / √6) ≈ 4.41

## Step 3: Decision
- Critical value (one-tailed, α = 0.05): t0.05,5 ≈ 2.015  
- Since 4.41 > 2.015, reject H0

## Conclusion
At the 5% significance level, there is sufficient evidence to conclude that the training program improved employee productivity.

Ans 8 # Chi-square Test of Independence: Product Preference vs Gender

## Step 1: Observed Data

| Gender  | Product A | Product B | Product C | Total |
|---------|-----------|-----------|-----------|-------|
| Male    | 20        | 30        | 25        | 75    |
| Female  | 30        | 25        | 20        | 75    |
| Total   | 50        | 55        | 45        | 150   |

## Step 2: Hypotheses
- H0: Product preference is independent of gender  
- H1: Product preference depends on gender  
- Significance level: α = 0.05

## Step 3: Expected Frequencies
E = (row total × column total) / grand total

- Male–Product A: 25  
- Male–Product B: 27.5  
- Male–Product C: 22.5  
- Female–Product A: 25  
- Female–Product B: 27.5  
- Female–Product C: 22.5  

## Step 4: Test Statistic
χ² = Σ (O − E)² / E  

= (20−25)²/25 + (30−27.5)²/27.5 + (25−22.5)²/22.5  
+ (30−25)²/25 + (25−27.5)²/27.5 + (20−22.5)²/22.5  

≈ 3.01

## Step 5: Degrees of Freedom
df = (rows−1)(columns−1) = (2−1)(3−1) = 2

Critical value at α = 0.05, df = 2 → 5.991

## Step 6: Decision
- Calculated χ² = 3.01  
- Critical value = 5.991  
- Since 3.01 < 5.991, fail to reject H0
