Question 3 : What is the difference between Normalization and Standardization?

Normalization squeezes all your data points into a specific, tiny range—usually between 0 and 1.

Standardization centers the data around an average (mean) of 0 with a "spread" (standard deviation) of 1.


Question 4 : A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when they should be used.

1. Imputation (Filling the Gaps)
Instead of throwing away data,  fill the missing spots with a calculated value. The most common approach is using the Mean (average) or Median (middle value) of the existing ages.

When to use it: Use this when the number of missing values is relatively small (e.g., less than 5–10%) and the data is mostly consistent.

Why: It allows you to keep the rest of the information in those rows (like the person's "Income" or "Location") without losing the entire record.

Use the Median if you have "outliers" (like a few 100-year-olds in a group of teenagers), as the average would get pulled too high.

Deletion (Dropping Rows)
This is the "clean slate" approach where you simply remove any row that is missing an age. This is often called Listwise Deletion.

Question 6 : What is One-Hot Encoding? Give an example with the categories: “Red, Blue, Green”.

A6.  One-Hot Encoding is a process used in data transformation to convert categorical data (words or labels) into a numerical format that machine learning algorithms can understand.

Most computers can't "do math" on the word "Red," but they can on the number 1. However, we can't just use 1, 2, and 3 because the computer might think "Green" (3) is "greater than" or "better than" "Red" (1). One-Hot Encoding solves this by creating a separate column for every category.

Question 8 : Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist

A8  We prefer Z-score Standardization over Min-Max Scaling when outliers exist because Min-Max Scaling is highly sensitive to extreme values; a single outlier can "squash" the rest of the data into a tiny, indistinguishable range (like 0.001 to 0.002) because it scales everything relative to the absolute minimum and maximum.

Z-score uses the mean and standard deviation, which allows it to center the data around the average and maintain the relative distribution of "normal" data points regardless of how extreme an outlier might be. This ensures that the majority of your data remains meaningful and doesn't lose its internal structure just because of a few freak data points.
