Question 2 : What is data extraction? Explain its role in the ETL pipeline.

A2. Data extraction is the process of retrieving raw data from various source systems (databases, files, APIs, spreadsheets, cloud storage, etc.) so it can be prepared for further processing.

It's the first stage of the ETL pipeline (Extract → Transform → Load).

The goal is to collect data in its original form without altering it, ensuring completeness and accuracy.

the role of this in ETL pipeline is to enure all relivent data is pulled out in the pipeline to process further.
