Q4. Why is median imputation preferred over mean imputation for skewed data such as income?

A4. Median imputation is preferred for skewed data like income because the mean is highly sensitive to extreme values (outliers), while the median is robust.

Q5. What is forward fill and in what type of dataset is it most useful?

A5. Forward fill is most effective—and most commonly used—in Time-Series Datasets. In these datasets, the order of observations matters, and data points are often related to the ones that immediately preceded them.
