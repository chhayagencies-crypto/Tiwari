Q1. explain the difference between Relative , Absolute and Mixed Cell Referencing.

A1. Relative Cell Referencing: Adjusts automatically when the formula is copied to another cell.
EX: =A1+B1,  if copied from row 1 to row 2, it becomes= A2+B2
    Absolute Cell Referencing:Remains fixed no matter where the formula is copied.
    EX: =$A$1 + $B$1
    Always refers to cells A1 and B1, even if moved or copied elsewhere.

    Mixed Cell Referencing: Part of the reference is fixed, and part adjusts.

    =A$1 → Column changes, row stays fixed.

=$A1 → Row changes, column stays fixed.

 Offers flexibility when copying formulas across rows or columns.
